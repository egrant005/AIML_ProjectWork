# 3. (Part 2) -  Remove duplicate movies, renumber Movie Ids & update Ratings DataFrame

```python
# Determine Duplicate movies based on title and release date to keep & remove
DUP_movies_first = movies.duplicated(subset=['movie title','release date'],keep='first')
DUP_movies_last = movies.duplicated(subset=['movie title','release date'],keep='last')
DUP_idx_keep = []
DUP_idx_rem = []
DUP_ID_keep = []
DUP_ID_rem = []

for ii in range(0,len(movies)):
    if DUP_movies_last[movies.index[ii]]:
        DUP_idx_keep.append(movies.index[ii])
        DUP_ID_keep.append(movies['movie id'][movies.index[ii]])
    
    if DUP_movies_first[movies.index[ii]]:
        DUP_idx_rem.append(movies.index[ii])        
        DUP_ID_rem.append(movies['movie id'][movies.index[ii]])
        
for ii in range(0,len(DUP_idx_keep)-1):
    for jj in range(ii+1,len(DUP_idx_keep)):
        if movies['movie title'][DUP_idx_keep[ii]] == movies['movie title'][DUP_idx_keep[jj]] and \
           movies['release date'][DUP_idx_keep[ii]] == movies['release date'][DUP_idx_keep[jj]]:
            DUP_idx_keep[jj] = DUP_idx_keep[ii]
            DUP_ID_keep[jj] = DUP_ID_keep[ii]

```

```python
#Movies IDs to REMOVE
movies.loc[DUP_idx_rem]
```

```python
#Movies/IDs to KEEP
movies.loc[DUP_idx_keep]
```

```python
#Drop movie entries that are duplicates
movies.drop(movies.index[DUP_idx_rem],inplace=True)
```

```python
#Update ratings so that duplicate movie IDs point back to only the one ID kept in MOVIES DataFrame
for ii in range(0,len(ratings)):
    for jj in range(0,len(DUP_ID_rem)):
        if ratings['movie id'][ratings.index[ii]] == DUP_ID_rem[jj]:
            ratings.loc[ratings.index[ii],('movie id')] = DUP_ID_keep[jj]
            print(f'Changing DUPLICATE movie id {DUP_ID_rem[jj]} to {DUP_ID_keep[jj]}')
            break
    
```

```python
# FIND MISSING MOVIE IDs in Movies DataFrame
missing_found = 0
for ii in range(0,len(ratings)):
    found = 0
    for jj in range(0,len(movies)):
        if ratings['movie id'][ratings.index[ii]] == movies['movie id'][movies.index[jj]]:
            found = 1
            break
            
    if found == 0:
        missing_found = 1
        print(f"Movie ID {ratings['movie id'][ratings.index[ii]]} was found in Ratings DF but not found in Movies DF")

if missing_found == 1:
    print("\nMissing movie IDs in MOVIES DataFrame indicated above")
else:
    print("\nNo missing movie IDs in MOVIES DataFrame")
        
# FIND MISSING MOVIE IDs in Ratings DataFrame
missing_found = 0
for ii in range(0,len(movies)):
    found = 0
    for jj in range(0,len(ratings)):
        if movies['movie id'][movies.index[ii]] == ratings['movie id'][ratings.index[jj]]:
            found = 1
            break
            
    if found == 0:
        missing_found = 1
        print(f"Movie ID {movies['movie id'][movies.index[ii]]} was found in Movies DF but not found in Ratings DF")
    
if missing_found == 1:
    print("\nMissing movie IDs in RATINGS DataFrame indicated above")
else:
    print("\nNo missing movie IDs in RATINGS DataFrame")    
```

```python
# Movie ID to Index

movieID_2_index = np.zeros([max(movies['movie id'])+1,1])
for ii in range(0,len(movies)):
    movieID_2_index[movies['movie id'][movies.index[ii]]] = movies.index[ii]

movieID_2_index
```

```python
# Number of ratings of a movie ID
CNT_rating = ratings.groupby(by=['movie id'])['rating'].count()
print(CNT_rating)
```

```python
# Average rating for a movie ID
AVG_rating = ratings.groupby(by=['movie id'])['rating'].mean()
print(AVG_rating)
```

```python
#Determine movies with MORE than 100 ratings
movie_idx = []
for ii in range(0,len(AVG_rating)):
    if CNT_rating[CNT_rating.index[ii]] < 100:
        movie_idx.append(CNT_rating.index[ii])
movie_idx
```

```python
#Grab top 25 movies that have more than 100 ratings
TOP_25 = AVG_rating.drop(movie_idx).sort_values().tail(25)
print(TOP_25)
```

```python
TOP_25.index[24]
```

```python
int(movieID_2_index[TOP_25.index[24]])
```

```python
TOP_25_data = []
rank = 0
for ii in range(24,-1,-1):
    rank += 1
    TOP_25_data.append([rank,TOP_25[TOP_25.index[ii]],CNT_rating[TOP_25.index[ii]],movies['movie title'][int(movieID_2_index[TOP_25.index[ii]])],movies['movie id'][int(movieID_2_index[TOP_25.index[ii]])]])
TOP_25_data
```
